```python
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
```

```python
boston = datasets.load_boston()
# print(boston.DESCR)
y = boston.target
X = boston.data[y<50,5]
y = boston.target[y<50]
```

```python
plt.scatter(X, y)
```

```python
from playML.model_selection import train_test_split
X_train, X_test, y_train, y_test =train_test_split(X, y)
X_train, X_test=X_train.reshape(-1,1), X_test.reshape(-1,1)
```

```python
from playML.metrics import plot_learning_curve
from sklearn.linear_model import LinearRegression
clf = LinearRegression()
plot_learning_curve(clf, X_train, y_train, X_test, y_test)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
def polynomialRegression(degree):
    return Pipeline([
        ("poly",PolynomialFeatures(degree=degree)),
        ("scaler",StandardScaler()),
        ("lin_reg", LinearRegression())
    ])
```

```python
poly_2 = polynomialRegression(2)
plot_learning_curve(poly_2, X_train, y_train, X_test, y_test)
poly_2 = polynomialRegression(10)
plot_learning_curve(poly_2, X_train, y_train, X_test, y_test)
poly_2 = polynomialRegression(100)
plot_learning_curve(poly_2, X_train, y_train, X_test, y_test)
```

```python
from sklearn.linear_model import Ridge
def RidgeRegression(degree,alpha=1):
    return Pipeline([
        ("poly",PolynomialFeatures(degree=degree)),
        ("scaler",StandardScaler()),
        ("ridge_reg", Ridge())
    ])
```

```python
poly_2 = RidgeRegression(2,1)
plot_learning_curve(poly_2, X_train, y_train, X_test, y_test)
poly_2 = RidgeRegression(10, 1)
plot_learning_curve(poly_2, X_train, y_train, X_test, y_test)
poly_2 = RidgeRegression(100, 1)
plot_learning_curve(poly_2, X_train, y_train, X_test, y_test)
```

```python
poly_2 = RidgeRegression(2,10)
plot_learning_curve(poly_2, X_train, y_train, X_test, y_test)
poly_2 = RidgeRegression(10,10)
plot_learning_curve(poly_2, X_train, y_train, X_test, y_test)
poly_2 = RidgeRegression(100,10)
plot_learning_curve(poly_2, X_train, y_train, X_test, y_test)
```

```python
from sklearn.linear_model import Lasso
def LassoRegression(degree,alpha=1):
    return Pipeline([
        ("poly",PolynomialFeatures(degree=degree)),
        ("scaler",StandardScaler()),
        ("lasso_reg", Lasso(alpha = alpha))
    ])
```

```python
poly_2 = LassoRegression(2,10)
plot_learning_curve(poly_2, X_train, y_train, X_test, y_test)
poly_2 = LassoRegression(10,10)
plot_learning_curve(poly_2, X_train, y_train, X_test, y_test)
poly_2 = LassoRegression(100,10)
plot_learning_curve(poly_2, X_train, y_train, X_test, y_test)
```
