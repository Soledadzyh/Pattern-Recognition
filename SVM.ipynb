```python
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
X = X[y<2,:2]
y = y[y<2]
```

```python
from sklearn.preprocessing import StandardScaler
standardscaler = StandardScaler()
X_std = standardscaler.fit(X).transform(X)
```

```python
from sklearn.svm import LinearSVC
svc = LinearSVC(C=1e9)
svc.fit(X_std, y)
```

```python
from playML.metrics import plot_decision_boundary
plot_decision_boundary(svc, X_std)
plt.scatter(X_std[y==0,0],
            X_std[y==0,1])
plt.scatter(X_std[y==1,0],X_std[y==1,1])
```

```python
from sklearn.svm import LinearSVC
svc = LinearSVC(C=0.01)
svc.fit(X_std, y)
plot_decision_boundary(svc, X_std)
plt.scatter(X_std[y==0,0],
            X_std[y==0,1])
plt.scatter(X_std[y==1,0],X_std[y==1,1])
```
